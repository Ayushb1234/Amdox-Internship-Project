```python
import os
import json
import pandas as pd


class DashboardKPI:

    def __init__(self):

        os.makedirs("../outputs/dashboard", exist_ok=True)

    ##########################################################

    def load_files(self):

        self.sales = pd.read_csv(
            r"C:\Users\ka843\Coding\Amdox Internship_project\data\clean\clean_sales.csv"
        )

        self.customer = pd.read_csv(
            r"C:\Users\ka843\Coding\Amdox Internship_project\src\outputs\customer\customer_360.csv"
        )

        self.inventory = pd.read_csv(
            r"C:\Users\ka843\Coding\Amdox Internship_project\src\outputs\inventory\product_inventory.csv"
        )

        self.metrics = pd.read_csv(
            r"C:\Users\ka843\Coding\Amdox Internship_project\src\outputs\forecasting\metrics\model_metrics.csv"
        )

    ##########################################################

    def build(self):

        best = self.metrics.sort_values("MAPE").iloc[0]

        kpi = {

            "Revenue":
            float(self.sales["Total_Price"].sum()),

            "Orders":
            int(self.sales["Invoice"].nunique()),

            "Customers":
            int(self.sales["Customer_ID"].nunique()),

            "Products":
            int(self.sales["StockCode"].nunique()),

            "Countries":
            int(self.sales["Country"].nunique()),

            "Forecast_Model":
            best["Model"],

            "Forecast_MAPE":
            float(best["MAPE"]),

            "High_Risk_Customers":
            int(
                (self.customer["Churn_Risk"]=="High").sum()
            ),

            "Urgent_Products":
            int(
                (self.inventory["Priority"]=="Urgent").sum()
            ),

            "Reorder_Products":
            int(
                (self.inventory["Priority"]=="Reorder").sum()
            )

        }

        with open(

            "../outputs/dashboard/dashboard_kpis.json",

            "w"

        ) as f:

            json.dump(

                kpi,

                f,

                indent=4

            )

        pd.DataFrame(

            [kpi]

        ).to_csv(

            "../outputs/dashboard/dashboard_kpis.csv",

            index=False

        )

        print("\nDashboard KPI Generated Successfully\n")

        print(pd.DataFrame([kpi]))

    ##########################################################

    def run(self):

        self.load_files()

        self.build()


if __name__=="__main__":

    DashboardKPI().run()
```

```python
import pandas as pd

file_path = r"C:\Users\ka843\Coding\Amdox Internship_project\data\raw\online_retail_II.xlsx"

# Read both sheets
df_2009 = pd.read_excel(file_path, sheet_name="Year 2009-2010")
df_2010 = pd.read_excel(file_path, sheet_name="Year 2010-2011")

# Combine them
df = pd.concat([df_2009, df_2010], ignore_index=True)

print(df.shape)
df.head()
```

```python
# to save this dataset to csv

df.to_csv(r"C:\Users\ka843\Coding\Amdox Internship_project\data\raw\online_retail_II_combined.csv", index=False)

```

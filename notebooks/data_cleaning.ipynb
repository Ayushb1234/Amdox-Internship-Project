```python
import pandas as pd


class DataCleaner:

    def __init__(self, filepath):
        self.filepath = filepath

    def clean(self):

        print("="*60)
        print("Loading Dataset...")
        print("="*60)

        df = pd.read_csv(self.filepath)

        print(f"Original Shape : {df.shape}")

        ###################################################
        # Rename Columns
        ###################################################

        df.columns = (
            df.columns
            .str.strip()
            .str.replace(" ", "_")
        )

        ###################################################
        # Convert InvoiceDate
        ###################################################

        df["InvoiceDate"] = pd.to_datetime(
            df["InvoiceDate"],
            errors="coerce"
        )

        ###################################################
        # Fill Missing Description
        ###################################################

        df["Description"] = df["Description"].fillna("Unknown")

        ###################################################
        # Create Total Price
        ###################################################

        df["Total_Price"] = df["Quantity"] * df["Price"]

        ###################################################
        # Remove Duplicate Rows
        ###################################################

        before = len(df)

        df = df.drop_duplicates()

        after = len(df)

        print(f"Duplicates Removed : {before-after}")

        ###################################################
        # Remove Invalid Dates
        ###################################################

        df = df[df["InvoiceDate"].notna()]

        ###################################################
        # Strip Spaces
        ###################################################

        df["Description"] = df["Description"].str.strip()

        ###################################################
        # Add Transaction Type
        ###################################################

        df["Transaction_Type"] = "Purchase"

        df.loc[df["Quantity"] < 0,
               "Transaction_Type"] = "Return"

        ###################################################
        # Create Positive Sales Dataset
        ###################################################

        sales_df = df[
            (df["Quantity"] > 0) &
            (df["Price"] > 0)
        ].copy()

        ###################################################
        # Customer Dataset
        ###################################################

        customer_df = sales_df[
            sales_df["Customer_ID"].notna()
        ].copy()

        ###################################################
        # Save Files
        ###################################################

        sales_df.to_csv(
            r"C:\Users\ka843\Coding\Amdox Internship_project\data\clean\clean_sales.csv",
            index=False
        )

        customer_df.to_csv(
            r"C:\Users\ka843\Coding\Amdox Internship_project\data\clean\customer_sales.csv",
            index=False
        )

        print()

        print("="*60)
        print("Cleaning Completed")
        print("="*60)

        print("Sales Shape :", sales_df.shape)
        print("Customer Shape :", customer_df.shape)

        print()

        print("Files Saved")

        print("clean_sales.csv")

        print("customer_sales.csv")


if __name__ == "__main__":

    cleaner = DataCleaner(
        r"C:\Users\ka843\Coding\Amdox Internship_project\data\raw\online_retail_II_combined.csv"
    )

    cleaner.clean()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

plt.style.use("ggplot")


class RetailEDA:

    def __init__(self, filepath):

        self.filepath = filepath
        self.df = None

        self.output_folder = "../outputs/eda"

        os.makedirs(self.output_folder, exist_ok=True)

    ##########################################################

    def load_data(self):

        print("=" * 60)
        print("Loading Clean Dataset")
        print("=" * 60)

        self.df = pd.read_csv(
            self.filepath,
            parse_dates=["InvoiceDate"]
        )

        print(self.df.shape)

    ##########################################################

    def business_summary(self):

        print("\nBUSINESS SUMMARY\n")

        print("Total Transactions :", self.df["Invoice"].nunique())

        print("Total Customers    :", self.df["Customer_ID"].nunique())

        print("Total Products     :", self.df["StockCode"].nunique())

        print("Countries          :", self.df["Country"].nunique())

        print("Revenue            : {:.2f}".format(
            self.df["Total_Price"].sum()
        ))

        print("Units Sold         :", self.df["Quantity"].sum())

    ##########################################################

    def create_time_columns(self):

        self.df["Year"] = self.df["InvoiceDate"].dt.year

        self.df["Month"] = self.df["InvoiceDate"].dt.month_name()

        self.df["Month_Number"] = self.df["InvoiceDate"].dt.month

        self.df["Day"] = self.df["InvoiceDate"].dt.day

        self.df["Weekday"] = self.df["InvoiceDate"].dt.day_name()

        self.df["Hour"] = self.df["InvoiceDate"].dt.hour

    ##########################################################

    def monthly_sales(self):

        monthly = (

            self.df

            .groupby(["Year", "Month_Number"])

            ["Total_Price"]

            .sum()

            .reset_index()

        )

        plt.figure(figsize=(12,5))

        plt.plot(
            monthly.index,
            monthly["Total_Price"],
            linewidth=3
        )

        plt.title("Monthly Revenue")

        plt.xlabel("Months")

        plt.ylabel("Revenue")

        plt.tight_layout()

        plt.savefig(
            self.output_folder +
            "/monthly_sales.png"
        )

        plt.close()

        print("Monthly Sales Chart Saved")

    ##########################################################

    def top_products(self):

        top = (

            self.df

            .groupby("Description")

            ["Total_Price"]

            .sum()

            .sort_values(ascending=False)

            .head(10)

        )

        plt.figure(figsize=(10,6))

        top.sort_values().plot(kind="barh")

        plt.title("Top 10 Products")

        plt.tight_layout()

        plt.savefig(
            self.output_folder +
            "/top_products.png"
        )

        plt.close()

        print("Top Product Chart Saved")

    ##########################################################

    def top_countries(self):

        top = (

            self.df

            .groupby("Country")

            ["Total_Price"]

            .sum()

            .sort_values(ascending=False)

            .head(10)

        )

        plt.figure(figsize=(10,6))

        top.sort_values().plot(kind="barh")

        plt.title("Top Countries")

        plt.tight_layout()

        plt.savefig(
            self.output_folder +
            "/top_countries.png"
        )

        plt.close()

        print("Top Countries Chart Saved")

    ##########################################################

    def weekday_sales(self):

        order = [

            "Monday",
            "Tuesday",
            "Wednesday",
            "Thursday",
            "Friday",
            "Saturday",
            "Sunday"

        ]

        sales = (

            self.df

            .groupby("Weekday")

            ["Total_Price"]

            .sum()

            .reindex(order)

        )

        plt.figure(figsize=(10,5))

        sales.plot(kind="bar")

        plt.title("Weekday Sales")

        plt.tight_layout()

        plt.savefig(
            self.output_folder +
            "/weekday_sales.png"
        )

        plt.close()

        print("Weekday Chart Saved")

    ##########################################################

    def hourly_sales(self):

        sales = (

            self.df

            .groupby("Hour")

            ["Total_Price"]

            .sum()

        )

        plt.figure(figsize=(10,5))

        plt.plot(
            sales.index,
            sales.values,
            linewidth=3
        )

        plt.title("Hourly Revenue")

        plt.tight_layout()

        plt.savefig(
            self.output_folder +
            "/hourly_sales.png"
        )

        plt.close()

        print("Hourly Sales Saved")

    ##########################################################

    def run(self):

        self.load_data()

        self.business_summary()

        self.create_time_columns()

        self.monthly_sales()

        self.top_products()

        self.top_countries()

        self.weekday_sales()

        self.hourly_sales()

        print("\nEDA Completed Successfully")


if __name__ == "__main__":

    eda = RetailEDA(
        r"C:\Users\ka843\Coding\Amdox Internship_project\data\clean\clean_sales.csv"
    )

    eda.run()
```

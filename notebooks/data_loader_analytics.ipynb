```python
import pandas as pd


class DataLoader:

    def __init__(self, filepath):
        self.filepath = filepath

    def load_data(self):
        """
        Load CSV dataset.
        """
        try:
            df = pd.read_csv(self.filepath)

            print("=" * 50)
            print("Dataset Loaded Successfully")
            print("=" * 50)

            print(f"Rows    : {df.shape[0]}")
            print(f"Columns : {df.shape[1]}")

            return df

        except Exception as e:
            print("Error Loading Dataset")
            print(e)
            return None


if __name__ == "__main__":

    loader = DataLoader(r"C:\Users\ka843\Coding\Amdox Internship_project\data\raw\online_retail_II_combined.csv")

    df = loader.load_data()

    if df is not None:
        print(df.head())
```

```python
import pandas as pd
import os


class ForecastFeatureBuilder:

    def __init__(self, filepath):

        self.filepath = filepath

        os.makedirs("../outputs/forecasting/datasets", exist_ok=True)

    ############################################################

    def load_data(self):

        print("=" * 60)
        print("Loading Daily Dataset")
        print("=" * 60)

        self.df = pd.read_csv(
            self.filepath,
            parse_dates=["Date"]
        )

        print(self.df.shape)

    ############################################################

    def lag_features(self):

        print("\nCreating Lag Features...")

        self.df["Lag_1"] = self.df["Revenue"].shift(1)

        self.df["Lag_7"] = self.df["Revenue"].shift(7)

        self.df["Lag_14"] = self.df["Revenue"].shift(14)

        self.df["Lag_30"] = self.df["Revenue"].shift(30)

    ############################################################

    def rolling_features(self):

        print("Creating Rolling Features...")

        self.df["Rolling_Mean_7"] = (

            self.df["Revenue"]

            .rolling(7)

            .mean()

        )

        self.df["Rolling_Mean_14"] = (

            self.df["Revenue"]

            .rolling(14)

            .mean()

        )

        self.df["Rolling_Mean_30"] = (

            self.df["Revenue"]

            .rolling(30)

            .mean()

        )

        self.df["Rolling_STD_7"] = (

            self.df["Revenue"]

            .rolling(7)

            .std()

        )

        self.df["Rolling_STD_30"] = (

            self.df["Revenue"]

            .rolling(30)

            .std()

        )

    ############################################################

    def growth_features(self):

        print("Creating Growth Features...")

        self.df["Revenue_Growth"] = (

            self.df["Revenue"]

            .pct_change()

        )

    ############################################################

    def clean(self):

        print("Removing Initial NaNs...")

        self.df = self.df.dropna()

        print(self.df.shape)

    ############################################################

    def save(self):

        self.df.to_csv(

            "../outputs/forecasting/datasets/forecast_features.csv",

            index=False

        )

        print("\nForecast Feature Dataset Saved")

    ############################################################

    def run(self):

        self.load_data()

        self.lag_features()

        self.rolling_features()

        self.growth_features()

        self.clean()

        self.save()

        print("\nFeature Dataset Ready")


if __name__ == "__main__":

    obj = ForecastFeatureBuilder(

        r"C:\Users\ka843\Coding\Amdox Internship_project\src\outputs\forecasting\datasets\daily_sales.csv"

    )

    obj.run()
```
